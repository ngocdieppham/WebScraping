```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page_link='http://phuong13021982.pythonanywhere.com/'
page_response = requests.get(page_link)
page_content = BeautifulSoup(page_response.content,"html.parser")

product_image = page_content.find_all('img',attrs={"class":"card-img-top"})
list_of_image =[]
for item in product_image:
    list_of_image.append(item['src'].split('/')[-1])
product_name = page_content.find_all('h4',attrs={"class":"card-title"})
list_of_name =[]
for item in product_name:
    list_of_name.append(item.text)
details = page_content.find_all('p',attrs={"class":"card-text"})
list_of_description =[]
list_of_price =[]
for item in details:
    #lấy text của item, chuyển <br/> thành \n và lưu vào biến components
    components = item.get_text(separator="\n") 
    components = components.split('\n')
    list_of_description.append(components[0])
    price = components[1].split(" vnđ")
    list_of_price.append(float(price[0].replace(',','')))
#print(list_of_image)
#print(list_of_name)
#print(list_of_description)
#print(list_of_price)
#print(len(list_of_image),len(list_of_name),len(list_of_description),len(list_of_price))
data = pd.DataFrame({"name":list_of_name,"description":list_of_description,"price":list_of_price,"image":list_of_image})
data
```
